##### Normalized Foot Traffic in January

- Gets  3 week rolling average of foot traffic at the end of Jan/beginning of Feb, which is then normalized to the number of devices seen in the panel and number of stores per zip code
- Normalizing to the number of devides seen in the panel accounts for growth/shrinking of recorded businesses
- Normalizing to the number of store per zip code ensures that foot traffic isn't high/low simply due to the number of businesses in that area

```python
# end of january/beginning of feb dates
jan_first_condition = hh_res['date_range_end'] == datetime.date(2020, 1, 27)
jan_second_condition = hh_res['date_range_end'] == datetime.date(2020, 2, 3)
jan_third_condition = hh_res['date_range_end'] == datetime.date(2020, 2, 10)
jan_week_visits = hh_res[(jan_first_condition) | (jan_second_condition) | (jan_third_condition)] # date range = 1/20 - 2/10

# find the number of visits in each zip code
jan_week_visits_per_zip = jan_week_visits.groupby('postal_code', as_index = False).sum()[['postal_code','raw_visit_counts']]
jan_week_visits_per_zip['raw_visit_counts'] = jan_week_visits_per_zip['raw_visit_counts']/3

# --- start normalization ---
# divide each foot count by the number of devices in the panel
jan_norm_stats_w1 = pd.read_csv('normalization_stats/2020-01-20-normalization-stats.csv')
jan_norm_stats_w2 = pd.read_csv('normalization_stats/2020-01-27-normalization-stats.csv')
jan_norm_stats_w3 = pd.read_csv('normalization_stats/2020-02-03-normalization-stats.csv')
agg_jan_norm_stats = sum(jan_norm_stats_w1['total_devices_seen'].astype(int)) + sum(jan_norm_stats_w2['total_devices_seen'].astype(int)) + sum(jan_norm_stats_w3['total_devices_seen'].astype(int))
jan_week_visits_per_zip['normalized_visit_counts'] = (jan_week_visits_per_zip['raw_visit_counts']/agg_jan_norm_stats)

# divide this by the number of stores in each zip code
jan_stores_counts = jan_week_visits.groupby('postal_code', as_index = False).count()[['postal_code','id']]
jan_week_visits_per_zip = pd.merge(jan_week_visits_per_zip, jan_stores_counts, on = "postal_code")
jan_week_visits_per_zip['normalized_visit_counts'] = (jan_week_visits_per_zip['normalized_visit_counts']/jan_week_visits_per_zip['id']) 
# --- end normalization ---

# adding a column to specify that this data is representative of January data
jan_week_visits_per_zip['month'] = "JAN"

print(jan_week_visits_per_zip.shape)
jan_week_visits_per_zip.head()
```

##### Normalized Foot Traffic in March

- Gets  3 week rolling average of foot traffic at the end of March/beginning of April (indicating the start of quarantine), which is then normalized to the number of devices seen in the panel and number of stores per zip code
- Normalizing to the number of devides seen in the panel accounts for growth/shrinking of recorded businesses
- Normalizing to the number of store per zip code ensures that foot traffic isn't high/low simply due to the number of businesses in that area

```python
# the week of the national emergency dates
nm_first_condition = hh_res['date_range_end'] == datetime.date(2020, 3, 23)
nm_second_condition = hh_res['date_range_end'] == datetime.date(2020, 3, 30)
nm_third_condition = hh_res['date_range_end'] == datetime.date(2020, 4, 6)
nat_emerg_week = hh_res[(nm_first_condition) | (nm_second_condition) | (nm_third_condition)] # range = 3/16 - 4/6

# find the number of visits in each zip code
nat_emerg_week_per_zip = nat_emerg_week.groupby('postal_code', as_index = False).sum()[['postal_code','raw_visit_counts']]
nat_emerg_week_per_zip['raw_visit_counts'] = nat_emerg_week_per_zip['raw_visit_counts']/3

# --- start normalization ---
# divide each foot count by the number of devices in the panel
nat_emerg_norm_stats_w1 = pd.read_csv('normalization_stats/2020-03-16-normalization-stats.csv') 
nat_emerg_norm_stats_w2 = pd.read_csv('normalization_stats/2020-03-23-normalization-stats.csv')
nat_emerg_norm_stats_w3 = pd.read_csv('normalization_stats/2020-03-30-normalization-stats.csv')
agg_nat_emerg_norm_stats = sum(nat_emerg_norm_stats_w1['total_devices_seen'].astype(int)) + sum(nat_emerg_norm_stats_w2['total_devices_seen'].astype(int)) + sum(nat_emerg_norm_stats_w3['total_devices_seen'].astype(int))
nat_emerg_week_per_zip['normalized_visit_counts'] = (nat_emerg_week_per_zip['raw_visit_counts']/agg_nat_emerg_norm_stats)

# divide this by the number of stores in each zip code
nat_emerg_stores_counts = nat_emerg_week.groupby('postal_code', as_index = False).count()[['postal_code','id']]
nat_emerg_week_per_zip = pd.merge(nat_emerg_week_per_zip, nat_emerg_stores_counts, on = "postal_code")
nat_emerg_week_per_zip['normalized_visit_counts'] = (nat_emerg_week_per_zip['normalized_visit_counts']/nat_emerg_week_per_zip['id'])
# --- end normalization ---

# adding a column to specify that this data is representative of national emergency data
nat_emerg_week_per_zip['month'] = "MAR"

print(nat_emerg_week_per_zip.shape)
nat_emerg_week_per_zip.head(2)
```

##### Normalized Foot Traffic in December

- Gets  3 week rolling average of foot traffic at the end of December (indicating the second shut down of CA), which is then normalized to the number of devices seen in the panel and number of stores per zip code
- Normalizing to the number of devides seen in the panel accounts for growth/shrinking of recorded businesses
- Normalizing to the number of store per zip code ensures that foot traffic isn't high/low simply due to the number of businesses in that area

```python
# the week of the second california shut down
dec_first_condition = hh_res['date_range_end'] == datetime.date(2020, 12, 14)
dec_second_condition = hh_res['date_range_end'] == datetime.date(2020, 12, 21)
dec_third_condition = hh_res['date_range_end'] == datetime.date(2020, 12, 28)
dec_week = hh_res[(dec_first_condition) | (dec_second_condition) | (dec_third_condition)] # range = 12/07 - 12/28

# find the number of visits in each zip code
dec_week_per_zip = dec_week.groupby('postal_code', as_index = False).sum()[['postal_code','raw_visit_counts']]
dec_week_per_zip['raw_visit_counts'] = dec_week_per_zip['raw_visit_counts']/3

# --- start normalization --- 
# divide each foot count by the number of devices in the panel
dec_norm_stats_w1 = pd.read_csv('normalization_stats/2020-12-14-normalization-stats.csv')
dec_norm_stats_w2 = pd.read_csv('normalization_stats/2020-12-21-normalization-stats.csv')
dec_norm_stats_w3 = pd.read_csv('normalization_stats/2020-12-28-normalization-stats.csv')
dec_norm_stats_w1 = dec_norm_stats_w1[dec_norm_stats_w1['region'] == 'ALL_STATES'] # to match jan and mar/apr
dec_norm_stats_w2 = dec_norm_stats_w2[dec_norm_stats_w2['region'] == 'ALL_STATES']
dec_norm_stats_w3 = dec_norm_stats_w3[dec_norm_stats_w3['region'] == 'ALL_STATES']
agg_dec_norm_stats = sum(dec_norm_stats_w1['total_devices_seen'].astype(int)) + sum(dec_norm_stats_w2['total_devices_seen'].astype(int)) + sum(dec_norm_stats_w3['total_devices_seen'].astype(int))
dec_week_per_zip['normalized_visit_counts'] = (dec_week_per_zip['raw_visit_counts']/agg_dec_norm_stats)
# --- start normalization ---

# divide this by the number of stores in each zip code
dec_week_stores_counts = dec_week.groupby('postal_code', as_index = False).count()[['postal_code','id']]
dec_week_per_zip = pd.merge(dec_week_per_zip, dec_week_stores_counts, on = "postal_code")
dec_week_per_zip['normalized_visit_counts'] = (dec_week_per_zip['normalized_visit_counts']/dec_week_per_zip['id'])

# adding a column to specify that this data is representative of december data
dec_week_per_zip['month'] = "DEC"

print(dec_week_per_zip.shape)
dec_week_per_zip.head(2)
```

##### Combining the DataFrames

This portion combines the above dataframes for the different months in order to standardize the normalized visit counts to each other. This makes the different months comparable and set up our analysis for percentage changes between Jan-Mar and Jan-Dec. 

```python
combined = pd.concat([jan_week_visits_per_zip, nat_emerg_week_per_zip, dec_week_per_zip])

norm_input = combined['normalized_visit_counts']
minimum = 0
maximum = max(norm_input)
combined['rescaled_normalized_visit_counts'] = ((norm_input - minimum)*100) / (maximum - minimum) #rescale the numbers to be between 0 and 100

combined.head()
```

```python
jan_week_visits_per_zip = combined[combined['month'] == 'JAN']
```

```python
nat_emerg_week_per_zip = combined[combined['month'] == 'MAR']
```

```python
dec_week_per_zip =  combined[combined['month'] == 'DEC']
```

```python
# get percentage change in foot traffic between january --> march and january --> december 

df = pd.DataFrame()
df['postal_code'] = jan_week_visits_per_zip['postal_code']
df['R1'] = jan_week_visits_per_zip['normalized_visit_counts']

nat_emerg_week_per_zip_relevant = nat_emerg_week_per_zip[['postal_code', 'normalized_visit_counts']]
dec_week_per_zip_relevant = dec_week_per_zip[['postal_code', 'normalized_visit_counts']]

df = df.merge(nat_emerg_week_per_zip_relevant, how = 'inner', on='postal_code').rename(columns={"normalized_visit_counts": "R2"})
df = df.merge(dec_week_per_zip_relevant, how = 'inner', on='postal_code').rename(columns={"normalized_visit_counts": "R3"})

df.head()
```

```python
# sanity check that all zip codes got added to the above table
```

```python
# get the percentage change between values

df['jan_to_march'] = df[['R2', 'R1']].apply(lambda row: (row.iloc[0]-row.iloc[1])/row.iloc[0]*100, axis=1)
df['jan_to_december'] = df[['R3', 'R1']].apply(lambda row: (row.iloc[0]-row.iloc[1])/row.iloc[0]*100, axis=1)
df
```

```python
# update the sd geozip file to reflect which zip codes have percentage changes
sd_geo_zip_updated = sd_geo_zip.merge(df, left_on = 'zip', right_on='postal_code')
sd_geo_zip_updated.head(3)
```

```python
def top_5_businesses(df):
    """This function gathers the top 5 businesses based on number of locations in a certain zip code"""
    df = df.copy() # create a copy of the dataframe so as not to change the original one
    top_5_businesses = {}
    for i in df['postal_code'].unique():
        df_filtered = df[df['postal_code'] == i]
        df_unique_locations_grouped = df_filtered.groupby('safegraph_place_id', as_index = False).count()[['safegraph_place_id','id']]
        merged_df = pd.merge(df, df_unique_locations_grouped, on = 'safegraph_place_id')
        merged_df = merged_df.sort_values(by='id_y', ascending=False)
        merged_df = merged_df[:5]
        display(merged_df)
        stores = list(merged_df['location_name'])
        count = list(merged_df['id_y'])
        for k in range(len(stores)):
            if i not in top_5_businesses.keys():
                top_5_businesses[i] = [stores[k], count[k]]
            else:
                top_5_businesses[i].append([stores[k], count[k]])
    return top_5_businesses
  
jan_week_visits = hh_res[hh_res['date_range_end'] == datetime.date(2020, 1, 27)]
top_5_businesses(jan_week_visits)
```
